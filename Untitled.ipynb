```python
import pandas as pd
import numpy as np
from os import listdir
```

```python
nomf = listdir("ressources")
```

```python
df = pd.DataFrame(index=range(len(nomf)), columns=['Nom_du_fichier','Contenu'])
df.head()
```

```python
for i, fname in enumerate(nomf):
    f = open('ressources/'+fname, 'r', encoding='utf-8')
    fcontent = f.readlines()
    f.close()
    
    df.at[i, "Nom_du_fichier"] = fname
    df.at[i, "Contenu"] = fcontent[0]
    
df.head()
```

```python
mois = ["Jan","Feb","Mar","Apr","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
```

```python
def SepareVille(v):
    for m in mois:
        if m in v:
            ville = v[:v.find(m)]
    return ville
```

```python
df["Ville"] = df["Nom_du_fichier"].apply(SepareVille)
df.head()
```

```python
def SepareDate(k):
    date = k.Nom_du_fichier[len(k.Ville):k.Nom_du_fichier.find(".txt")]
    return date
```

```python
df["Date"] = df.apply(SepareDate, axis=1)
df.head()
```

```python
df.dtypes
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%b%d_%Y")
df.head()
```

```python
df["Annee"] = df.Date.dt.year
df["Mois"] = df.Date.dt.month
df["Jour_du_mois"] = df.Date.dt.day
df["Jour_Semaine"] = df.Date.dt.day_of_week
df.head()
```

```python
df["Weekend"] = np.where((df.Jour_Semaine == 5) | (df.Jour_Semaine == 6), "Oui","Non")
df
```
